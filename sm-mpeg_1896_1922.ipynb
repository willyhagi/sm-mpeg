# Comparação Climatológica
### Normais Climatológicas do Serviço Meteorológico do Museu Paraense Emílio Goeldi (SM-MPEG; 1896-1922) e do Instituto Nacional de Meteorologia (INMET; 1981-2010) para Belém (PA). 

* Fontes: [Cunha e Bastos (1973)](https://repositorio.museu-goeldi.br/jspui/handle/mgoeldi/900) e [INMET](https://portal.inmet.gov.br/normais).

* Autor do código: Meteorol. Me. [Willy Hagi](http://taggo.one/willyhagi).  
   - CREA-AM 32258
   - RNP 0419347917

* Repositório: SM-MPEG no Github.

* **Observações**

   - Atenção para as unidades! A tabela original de Cunha e Bastos (1973) reporta os registros de Pressão Atmosférica em unidades de milímetros de mercúrio (mmHg), embora tenham sido publicadas como se estivessem em hectopascal (hPa);
   - Os registros de Direção de Vento originais estão distribuídos em categorias, mas foram convertidos para unidades de grau para a normal climatológica;
   - Os registros de Nebulosidade originais não estão em unidades de decimais, mas foram convertidos para a normal climatológica.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

# Abrir Tabela de Dados

```python
sm_mpeg = pd.read_excel('smmpeg_1896_1922.xlsx', sheet_name=1, skiprows=1)
inmet = pd.read_excel('smmpeg_1896_1922.xlsx', sheet_name=2, skiprows=1)
```

# Comparação de Normais

```python
plt.style.use('seaborn-talk')
```

* Variáveis:
   - Pressão Atmosférica;
   - Temperatura Média;
   - Temperatura Máxima Média;
   - Temperatura Mínima Média;
   - Temperatura Máxima Absoluta;
   - Temperatura Mínima Absoluta.
   
Normais do SM-MPEG **sempre** representadas em linhas pretas marcadas por quadrados e as do INMET por linhas pretas marcadas por asteriscos.

```python
fig, ax = plt.subplots(figsize=(18, 9), nrows=2, ncols=3, sharex=True)

nmeses = np.arange(1, 13, 1)
meses = ['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',
         'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ']

# a) pressão atmosférica
ax[0, 0].plot(nmeses, sm_mpeg['Pressão Atmosférica (hPa)'], color='black', marker='s')
ax[0, 0].plot(nmeses, inmet['Pressão Atmosférica (hPa)'], color='black', marker='*')
ax[0, 0].fill_between(nmeses, sm_mpeg['Pressão Atmosférica (hPa)'], inmet['Pressão Atmosférica (hPa)'],
                      color='grey', alpha=0.25)
ax[0, 0].set(ylabel='Pressão Atmosférica (hPa)')

# b) temperatura média
ax[0, 1].plot(nmeses, sm_mpeg['Temperatura Média (°C)'], color='black', marker='s')
ax[0, 1].plot(nmeses, inmet['Temperatura Média (°C)'], color='black', marker='*')
ax[0, 1].fill_between(nmeses, sm_mpeg['Temperatura Média (°C)'], inmet['Temperatura Média (°C)'],
                      color='grey', alpha=0.25)
ax[0, 1].set(ylabel='Temperatura Média (°C)')

# c) Temperatura Máxima Média (°C)
ax[0, 2].plot(nmeses, sm_mpeg['Temperatura Máxima Média (°C)'], color='black', marker='s')
ax[0, 2].plot(nmeses, inmet['Temperatura Máxima Média (°C)'], color='black', marker='*')
ax[0, 2].fill_between(nmeses, sm_mpeg['Temperatura Máxima Média (°C)'], inmet['Temperatura Máxima Média (°C)'],
                      color='grey', alpha=0.25)
ax[0, 2].set(ylabel='Temperatura Máxima Média (°C)')

# d) Temperatura Mínima Média (°C)
ax[1, 0].plot(nmeses, sm_mpeg['Temperatura Mínima Média (°C)'], color='black', marker='s')
ax[1, 0].plot(nmeses, inmet['Temperatura Mínima Média (°C)'], color='black', marker='*')
ax[1, 0].fill_between(nmeses, sm_mpeg['Temperatura Mínima Média (°C)'], inmet['Temperatura Mínima Média (°C)'],
                      color='grey', alpha=0.25)
ax[1, 0].set(ylabel='Temperatura Mínima Média (°C)')

# e) Temperatura Máxima Absoluta (°C)
ax[1, 1].plot(nmeses, sm_mpeg['Temperatura Máxima Absoluta (°C)'], color='black', marker='s')
ax[1, 1].plot(nmeses, inmet['Temperatura Máxima Absoluta (°C)'], color='black', marker='*')
ax[1, 1].fill_between(nmeses, sm_mpeg['Temperatura Máxima Absoluta (°C)'], inmet['Temperatura Máxima Absoluta (°C)'],
                      color='grey', alpha=0.25)
ax[1, 1].set(ylabel='Temperatura Máxima Absoluta (°C)')

# f) Temperatura Mínima Absoluta (°C)
ax[1, 2].plot(nmeses, sm_mpeg['Temperatura Mínima Absoluta (°C)'], color='black', marker='s')
ax[1, 2].plot(nmeses, inmet['Temperatura Mínima Absoluta (°C)'], color='black', marker='*')
ax[1, 2].fill_between(nmeses, sm_mpeg['Temperatura Mínima Absoluta (°C)'], inmet['Temperatura Mínima Absoluta (°C)'],
                      color='grey', alpha=0.25)
ax[1, 2].set(ylabel='Temperatura Mínima Absoluta (°C)')

ax[1, 2].set_xticks(nmeses)
ax[1, 2].set_xticklabels(meses)

plt.tight_layout()
plt.show()
```

* Variáveis:
   - Precipitação;
   - Precipitação Máxima em 24 Horas;
   - Umidade Relativa;
   - Dias de Chuva.

```python
fig, ax = plt.subplots(figsize=(15, 9), nrows=2, ncols=2, sharex=True)

nmeses = np.arange(1, 13, 1)

# a) Precipitação (mm)
ax[0, 0].plot(nmeses, sm_mpeg['Precipitação (mm)'], color='black', marker='s')
ax[0, 0].plot(nmeses, inmet['Precipitação (mm)'], color='black', marker='*')
ax[0, 0].fill_between(nmeses, sm_mpeg['Precipitação (mm)'], inmet['Precipitação (mm)'],
                      color='grey', alpha=0.25)
ax[0, 0].set(ylabel='Precipitação (mm)')

# b) Precipitação Máxima em 24 Horas (mm)
ax[0, 1].plot(nmeses, sm_mpeg['Precipitação Máxima em 24 Horas (mm)'], color='black', marker='s')
ax[0, 1].plot(nmeses, inmet['Precipitação Máxima em 24 Horas (mm)'], color='black', marker='*')
ax[0, 1].fill_between(nmeses, sm_mpeg['Precipitação Máxima em 24 Horas (mm)'],
                      inmet['Precipitação Máxima em 24 Horas (mm)'],
                      color='grey', alpha=0.25)
ax[0, 1].set(ylabel='Precipitação Máxima em 24 Horas (mm)')

# c) Umidade Relativa (%)
ax[1, 0].plot(nmeses, sm_mpeg['Umidade Relativa (%)'], color='black', marker='s')
ax[1, 0].plot(nmeses, inmet['Umidade Relativa (%)'], color='black', marker='*')
ax[1, 0].fill_between(nmeses, sm_mpeg['Umidade Relativa (%)'], inmet['Umidade Relativa (%)'],
                      color='grey', alpha=0.25)
ax[1, 0].set(ylabel='Umidade Relativa (\%)')

# d) Dias de Chuva
ax[1, 1].plot(nmeses, sm_mpeg['Dias de Chuva'], color='black', marker='s')
ax[1, 1].plot(nmeses, inmet['Dias de Chuva'], color='black', marker='*')
ax[1, 1].fill_between(nmeses, sm_mpeg['Dias de Chuva'], inmet['Dias de Chuva'],
                      color='grey', alpha=0.25)
ax[1, 1].set(ylabel='Dias de Chuva')

ax[1, 1].set_xticks(nmeses)
ax[1, 1].set_xticklabels(meses)


plt.tight_layout()
plt.show()
```

* Variáveis:
   - Direção do Vento;
   - Velocidade do Vento;
   - Nebulosidade.

```python
fig, ax = plt.subplots(figsize=(15, 9), nrows=3, ncols=1, sharex=True)

nmeses = np.arange(1, 13, 1)

# a) Direção do Vento
ax[0].plot(nmeses, sm_mpeg['Direção do Vento (°)'], color='black', marker='s')
ax[0].plot(nmeses, inmet['Direção do Vento (°)'], color='black', marker='*')
ax[0].fill_between(nmeses, sm_mpeg['Direção do Vento (°)'], inmet['Direção do Vento (°)'],
                      color='grey', alpha=0.25)
ax[0].set(ylabel='Direção do Vento (°)')


# b) Velocidade do Vento
ax[1].plot(nmeses, sm_mpeg['Velocidade do Vento (m/s)'], color='black', marker='s')
ax[1].plot(nmeses, inmet['Velocidade do Vento (m/s)'], color='black', marker='*')
ax[1].fill_between(nmeses, sm_mpeg['Velocidade do Vento (m/s)'],
                   inmet['Velocidade do Vento (m/s)'],
                   color='grey', alpha=0.25)
ax[1].set(ylabel='Velocidade do Vento (m/s)')


# c) Nebulosidade
ax[2].plot(nmeses, sm_mpeg['Nebulosidade'] / 10., color='black', marker='s')
ax[2].plot(nmeses, inmet['Nebulosidade'], color='black', marker='*')
ax[2].fill_between(nmeses, sm_mpeg['Nebulosidade'] / 10.,
                   inmet['Nebulosidade'],
                   color='grey', alpha=0.25)
ax[2].set(ylabel='Nebulosidade')

ax[2].set_xticks(nmeses)
ax[2].set_xticklabels(meses)

plt.tight_layout()
plt.show()
```
